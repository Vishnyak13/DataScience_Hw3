## Друга частина

Реалізуйте програму, яка буде будувати візерунок Серпінського. Історія дослідження цього візерунка досить довга, а сам візерунок представляє особливий інтерес у такій галузі, як геометрія фракталів. Візерунок Серпінського є випадковим об'єктом, визначеним рекурсивно, але в межі його форма прямує до детермінованого об'єкта.

Алгоритм побудови наступний:

Припустимо, що у вихідному стані ми маємо три точки на площині, причому їх положення описано координатами в деякій відповідній системі координат: 
1. Випадково вибирається деяка точка всередині трикутника, утвореного заданими вершинами.
2. Випадково вибирається одна із трьох вершин.
3. Вибирається точка, рівновіддалена від першої обраної точки та від обраної вершини.
4. Нова точка включається до зображення
5. Вихідна точка замінюється цією новою точкою із зображення.
6. Повторюється вся процедура, починаючи з кроку 2.

```python
import matplotlib.pyplot as plt
import random
```

```python
vertices = [(0, 0), (1, 0), (0.5, 1)]
current_point = (random.uniform(0, 1), random.uniform(0, 1))
points = [current_point]
num_points = 50000

for i in range(num_points):
    vertex = random.choice(vertices)
    new_point = ((current_point[0] + vertex[0]) / 2, (current_point[1] + vertex[1]) / 2)
    points.append(new_point)
    current_point = new_point

x, y = zip(*points)
plt.figure(figsize=(10, 10))
plt.scatter(x, y, s=0.1)
plt.title('Sierpinski pattern')
plt.show()
```
