```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./bestsellers with categories.csv')
df.head()
```

```python
df.shape
```

```python
sns.set_style('darkgrid')
plt.figure(figsize =(12,7))
sns.stripplot(x='Year', y='Reviews', hue='Genre', data=df)
plt.title('Розподіл відгуків в залежності від жанру та року')
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize =(12,7))
sns.boxenplot(x='Genre', y='User Rating', data=df)
plt.title('Графік рейтингу користувачів в залежності від жанру')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.violinplot(x='Year', y='Price', data=df)
plt.title('Розподіл цін в залежності від року видання')
plt.show()
```
