Побудуйте лінії регресії за допомогою функції regplot, залежності прибутковості однієї акції від іншої. Всього три графіки. Виберіть пару акцій, які найменш залежні в прибутковостях одна від одної.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.DataFrame(data = [[1, 25, 0,10], [2, -10, 15, 25],  [3, 10, -5, -15],
                                [4, 5, 5, -5], [5, 35, 20, -5], [6, 13, 25, 15]], 
                        columns = ['Момент спостереження', 'Прибутковість A, %', 'Прибутковість B, %', 'Прибутковість C, %'])
data
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Залежність прибутковості акцій', fontsize=20, y=1.02)
sns.regplot(x="Прибутковість A, %", y="Прибутковість B, %", data=data, ax=axes[0])\
            .set(title='Залежність прибутковості A vs B');
sns.regplot(x="Прибутковість A, %", y="Прибутковість C, %", data=data, ax=axes[1])\
            .set(title='Залежність прибутковості A vs C');
sns.regplot(x="Прибутковість B, %", y="Прибутковість C, %", data=data, ax=axes[2])\
            .set(title='Залежність прибутковості B vs C');
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(16,10))
sns.regplot(x="Прибутковість A, %", y="Прибутковість B, %", data=data, label='Прибутковість A vs B')
sns.regplot(x="Прибутковість A, %", y="Прибутковість C, %", data=data, label='Прибутковість A vs C')
sns.regplot(x="Прибутковість B, %", y="Прибутковість C, %", data=data, label='Прибутковість B vs C')
plt.title('Залежність прибутковості акцій')
plt.legend();
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data.corr(), annot=True, cmap="mako")
plt.title('Теплова карта прибутковості акцій');
```
